```python
df2.drop('Age', axis=1, inplace=True)
df2.drop('Gender', axis=1, inplace=True)
df2.drop('Total Bilirubin', axis=1, inplace=True)
df2.drop('Direct Bilirubin', axis=1, inplace=True)
df2.drop('Alkaline', axis=1, inplace=True)
df2.drop('Alamine', axis=1, inplace=True)
df2.drop('Asparte', axis=1, inplace=True)
df2.drop('Total Proteins', axis=1, inplace=True)
df2.drop('Albumin', axis=1, inplace=True)
df2.drop('Albumin and Globulin Ratio', axis=1, inplace=True)
df.drop('Liver or Non Liver', axis=1, inplace=True)
df #df2 dataframe'imizde sadece 1 sütun yani liver or non liver değeri bırakıldı, df dataframe'imizde ise 
#o sütun hariç hepsi bırakıldı
```

```python
df2
```

```python
x = df.values
x
```

```python
y = df2.values
y
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=0) #veri test splite sokuldu
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
knn_model = KNeighborsClassifier(n_neighbors = 11) 
knn_model.fit(x_train,np.ravel(y_train)) #KNN İÇİN GEREKEN KULLANILACAK KOMŞU SAYISI 11 OLARAK BELİRLENDİ.
predicted_y = knn_model.predict(x_test)
accuracy_knn = knn_model.score(x_test,y_test)#TAHMİN EDİLEN PREDICTED Y DEĞERİ X_TEST İLE GÖSTERİLDİ
print("KNN accuracy:",accuracy_knn)#ACCURACY BELİRLENMESİ İÇİN X_TEST VE Y_TEST KULLANILDI
```

```python
print(classification_report(y_test, predicted_y))
```

```python
logr = LogisticRegression(random_state=0)
logr.fit(x_train,np.ravel(y_train))
y_pred = logr.predict(x_test)
confusion_matrix(y_test,y_pred)
```

```python
accuracy_knn = logr.score(x_test,y_test)
print("Logistic Regression accuracy:",accuracy_knn)
#MODEL LOGISTIC REGRESSION ILE DE DENENDİ
```

```python
dtc = DecisionTreeClassifier(criterion = "entropy")
dtc.fit(x_train,np.ravel(y_train))
y_pred = dtc.predict(x_test)
cm=confusion_matrix(y_test,y_pred) #MODEL KARAR AĞACI İLE DENENDİ
print("Decision Tree")
print(cm)
```

```python
rfc = RandomForestClassifier(n_estimators = 10,criterion = 'entropy')
rfc.fit(x_train,np.ravel(y_train))
y_pred = rfc.predict(x_test)
cm = confusion_matrix(y_test,y_pred) #MODEL RASSAL ORMAN İLE DE DENENDİ.
print('Random Forest')
print(cm)
```
